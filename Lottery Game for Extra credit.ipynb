```python
import random
Money=200
a={'First prize':(0,10),'Second prize':(10,40),'Third prize':(40,80),'Thanks for join':(80,200)}

while True:
    print('input 1 to start next round')
    print('input 2 to stop game')
    while True:
        try:
            Q=input('Chose an option:')
            Q=int(Q)
            break
        except:
            print('Invalid input')
    print()

    if Q == 1:
        b = random.randint(0, 200)
        for x,v in a.items():
            if v[0]<=b<=v[1]:
                Money-=10
                if 0<b<=10:
                    Money+=100
                    print('Congratulate! You get:', str(x), 'Now you have:', Money, '$')
                elif 10<b<=40:
                    Money+=50
                    print('Congratulate! You get:', str(x), 'Now you have:', Money, '$')
                elif 40<b<=80:
                    Money+=10
                    print('Congratulate! You get:', str(x), 'Now you have:', Money, '$')
                elif 80<b<=200:
                    Money+=0
                    print('You get:',str(x),'Now you have:',Money,'$')
            if Money<=0:
                print('Game over!')
                break

    elif Q == 2:
        print('Quit Successfully!')
        break
```
